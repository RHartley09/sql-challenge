```python
import pandas as pd
import sqlalchemy as db
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect
import config
import matplotlib.pyplot as plt
```

```python
engine = db.create_engine(config.connection)
```

```python
conn = engine.connect()
```

```python
session = Session(engine)
```

```python
Base = automap_base()
```

```python
Base.prepare(engine, reflect=True)
```

```python
# I"m not sure why 'Base.classes.keys()' only gives me three tables.
Base.classes.keys()
```

```python
inspector_gadget = inspect(engine)
```

```python
# 'inspect' shows all tables.
inspector_gadget.get_table_names()
```

```python
salary_df = pd.read_sql("SELECT * FROM salaries", conn)
salary_df
```

```python
salrange = salary_df["salary"]
plt.hist(salrange)
plt.title("Salary Ranges")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.tight_layout
plt.show()
```

```python
titles_df = pd.read_sql("SELECT * FROM titles", conn)
titles_df
```

```python
dept_emp_df = pd.read_sql("SELECT * FROM dept_emp", conn)
dept_emp_df
```

```python
employees_df = pd.read_sql("SELECT * FROM employees", conn)
employees_df
```

```python
step1_df = pd.merge(salary_df, employees_df, on="emp_no")
step1_df.rename(columns = {'emp_title':'title_id'}, inplace = True)
step1_df.head()
```

```python
step2_df = pd.merge(step1_df, titles_df, on="title_id")
step2_df.head()
```

```python
salary_gb = step2_df.groupby(["title"], as_index=True)
salary_gb.head()
```

```python
salmean = salary_gb['salary'].mean()
salmean
```

```python
salmean.plot(kind="bar")
plt.tight_layout
plt.title("Avg Salary by Title")
plt.xlabel("Job Titles")
plt.ylabel("Salaries")
plt.show()
```
